```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
K = np.complex(250,10)
k = np.complex(15,2)
E_p = 4*K

def create_Ising_finite_set():
    N = 10
    U = create(N)
    H_initial = U.dag()*H_not*U
    
def time_evolution():
    t = np.linspace(0,10,1000)
    x_vec = np.linspace()
        
def create_cat_state():
    N = 10
    a = destroy(10)
    a_dag = a.dag()
    r_not = (((4*E_p*E_p) - ((k*k)/4))/(4*K*K))**(1/4)
    theta_not = 0.5*np.arctan((k/(((16*E_p*E_p) - k*k)**(0.5))))
    a_not = r_not * np.exp(np.complex(0,theta_not))
    
    H_not = (-1)*K*a_dag*a_dag*a*a + (E_p*a_dag*a_dag + E_p.conjugate()*a*a)
    H_eff = H_not - np.complex(0,1)*(k*a_dag*a*0.5)
    D = displace(N,a_not)
    
    H_eff_prime = D.dag()*H_eff*D
    s = coherent(N,a_not)
    #r = mesolve(H_eff_prime, s, tlist, [], [])
    return H_eff_prime*H_eff_prime.dag()

def create_finite_degenerate_state():
    N = 100
    U_not = create(N)
    U_not_trans = U.conj()
    amplitude_value = np.complex(4,2)
    initial_degenerate_state = coherent(N,amplitude_value)
    H_displaced = displace(N,a_not)*H*displace(N,a_not).dag()
    density_matrix = H_displaced*H_displaced.dag().unit()
    density_matrix.value[n] = density_matrix.unit().normalize()
    for i in enumerate(normalize(density_matrix)):
        create_cat_state(i.hamiltonian())
        
    for hamiltionain_set in i.hamiltonian():
        s = set(i.hamiltonian.dag())
        s = Qobj([N][N])
        liimit = s.dims()
        
    
def plot_wigner(r):
    figure, axis = plt.subplots(1, 1, figsize=(8,8))
    x = np.linspace(-7.5,7.5,200)
    W = wigner(r, x, x)
    #rint(W)
    limit = abs(W).max()
    axis.contourf(x, x, W, 1000, norm=mpl.colors.Normalize(-limit,limit), cmap=mpl.cm.get_cmap('RdBu'))
    #rint(abs(W).max())
    return figure, axis

```
